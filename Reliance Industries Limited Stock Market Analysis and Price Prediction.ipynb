```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from newsapi import NewsApiClient
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 0) Download VADER lexicon
nltk.download('vader_lexicon')

# 1) Fetch & prepare stock price data
symbol    = "RELIANCE.NS"
df_price  = yf.download(symbol, start="2023-01-01", end=datetime.today().strftime('%Y-%m-%d'))
df_price.columns = ['_'.join(c) if isinstance(c, tuple) else c for c in df_price.columns]
df_price = (
    df_price
    .reset_index()[['Date', f'Close_{symbol}', f'Volume_{symbol}']]
    .rename(columns={f'Close_{symbol}':'Close', f'Volume_{symbol}':'Volume'})
)
df_price['Date'] = pd.to_datetime(df_price['Date'])

# 2) Fetch & score news sentiment (last 30 days)
newsapi   = NewsApiClient(api_key='7b20751a6c044d668d12c26c03f4f1d8')
to_date   = datetime.today().strftime('%Y-%m-%d')
from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')

arts = newsapi.get_everything(
    q          = "Reliance Industries",
    from_param = from_date,
    to         = to_date,
    language   = 'en',
    sort_by    = 'relevancy',
    page_size  = 100
)['articles']

news_df = pd.DataFrame([{'date':a['publishedAt'][:10],'title':a['title']} for a in arts])
sid = SentimentIntensityAnalyzer()
news_df['sentiment'] = news_df['title'].map(lambda t: sid.polarity_scores(t)['compound'])
sentiment_daily = news_df.groupby('date')['sentiment'].mean().reset_index()
sentiment_daily['date'] = pd.to_datetime(sentiment_daily['date'])

# 3) Merge price + sentiment
df = (
    df_price
    .merge(sentiment_daily, how='left', left_on='Date', right_on='date')
    .drop(columns='date')
    .fillna(0)
    .sort_values('Date')
    .reset_index(drop=True)
)

# 4) Feature engineering
# Lag Features
df['lag_Close']     = df['Close'].shift(1)            # Previous day's close price
df['lag_Volume']    = df['Volume'].shift(1)           # Previous day's volume
df['lag_Sentiment'] = df['sentiment'].shift(1)        # Previous day's sentiment score

# Return & Change Features
df['Daily_Return']  = df['Close'].pct_change()        # Daily return percentage
df['Volume_Change'] = df['Volume'].pct_change()       # Daily volume change percentage

# Moving Averages
df['MA_5']          = df['Close'].rolling(5).mean()   # 5-day moving average
df['MA_10']         = df['Close'].rolling(10).mean()  # 10-day moving average
df['MA_20']         = df['Close'].rolling(20).mean()  # 20-day moving average (new added)

# Volatility (Standard Deviation)
df['Volatility_5']  = df['Daily_Return'].rolling(5).std()  # 5-day return volatility
df['Volatility_10'] = df['Daily_Return'].rolling(10).std() # 10-day return volatility (new added)

# Date-Based Features
df['Date'] = pd.to_datetime(df['Date'])              # Ensure 'Date' is datetime type
df['DayOfWeek'] = df['Date'].dt.dayofweek            # Day of the week (0=Monday, 6=Sunday)
df['Month']     = df['Date'].dt.month                # Month number (1 to 12)

# After creating features
df.replace([np.inf, -np.inf], np.nan, inplace=True)  # Replace inf with NaN
df.dropna(inplace=True)                             # Drop all rows with NaN
df.reset_index(drop=True, inplace=True)  

# Drop rows with NaN caused by rolling or shifting
df = df.dropna().reset_index(drop=True)

# 5) Prepare train/test
features = [
    'Volume', 'sentiment',
    'lag_Close', 'lag_Volume', 'lag_Sentiment',
    'Daily_Return', 'Volume_Change',
    'MA_5', 'MA_10', 'MA_20',
    'Volatility_5', 'Volatility_10',
    'DayOfWeek', 'Month'
]
X = df[features]
y = df['Close']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1, shuffle=False
)

# 6) Hyperparameter grid and models
n_estimators_list = [50, 100, 200, 300, 500]
model_classes = {
    'RandomForest': RandomForestRegressor,
    'XGBoost'      : XGBRegressor,
    'LightGBM'     : LGBMRegressor,
    'GradientBoost': GradientBoostingRegressor
}

def eval_model(mdl, X_test, y_test):
    pred = mdl.predict(X_test)
    return (
        mean_squared_error(y_test, pred),
        mean_absolute_error(y_test, pred),
        r2_score(y_test, pred)
    )

# 7) Add SVR (with scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
svr_C_list = [0.1, 1, 10]
svr_gamma_list = [0.01, 0.1]

# 8) Add LSTM (with reshaping)
def create_lstm_sequences(X, y, window=1):
    Xs, ys = [], []
    for i in range(len(X)-window):
        Xs.append(X.iloc[i:(i+window)].values)
        ys.append(y.iloc[i+window])
    return np.array(Xs), np.array(ys)

lstm_window = 1
X_lstm, y_lstm = create_lstm_sequences(X, y, window=lstm_window)
split_idx = int(len(X_lstm)*0.8)
X_lstm_train, X_lstm_test = X_lstm[:split_idx], X_lstm[split_idx:]
y_lstm_train, y_lstm_test = y_lstm[:split_idx], y_lstm[split_idx:]

# 9) Run all models and collect results
grid_results = []

# Tree-based models
for name, cls in model_classes.items():
    for n in n_estimators_list:
        if name == 'XGBoost':
            mdl = cls(n_estimators=n, random_state=1, verbosity=0)
        else:
            mdl = cls(n_estimators=n, random_state=1)
        mdl.fit(X_train, y_train)
        mse, mae, r2 = eval_model(mdl, X_test, y_test)
        grid_results.append([name, n, None, None, mse, mae, r2, mdl])

# SVR
for c in svr_C_list:
    for gamma in svr_gamma_list:
        svr = SVR(kernel='rbf', C=c, gamma=gamma)
        svr.fit(X_train_scaled, y_train)
        mse, mae, r2 = eval_model(svr, X_test_scaled, y_test)
        grid_results.append(['SVR', None, c, gamma, mse, mae, r2, svr])

# LSTM (single run, as deep learning grid search is slow)
lstm_model = Sequential()
lstm_model.add(LSTM(64, input_shape=(X_lstm_train.shape[1], X_lstm_train.shape[2])))
lstm_model.add(Dense(1))
lstm_model.compile(loss='mse', optimizer='adam')
lstm_model.fit(X_lstm_train, y_lstm_train, epochs=30, batch_size=8, verbose=0)
lstm_pred = lstm_model.predict(X_lstm_test).flatten()
lstm_mse = mean_squared_error(y_lstm_test, lstm_pred)
lstm_mae = mean_absolute_error(y_lstm_test, lstm_pred)
lstm_r2 = r2_score(y_lstm_test, lstm_pred)
grid_results.append(['LSTM', None, None, None, lstm_mse, lstm_mae, lstm_r2, lstm_model])

# 10) Results DataFrame
grid_df = pd.DataFrame(grid_results, columns=['Model','n_estimators','C','gamma','MSE','MAE','R2','_mdl'])
grid_df[['MSE','MAE','R2']] = grid_df[['MSE','MAE','R2']].round(3)
print("\n=== All model √ó estimator results ===")
print(grid_df[['Model','n_estimators','C','gamma','MSE','MAE','R2']])

# 11) Best per model
def get_best_row(group):
    return group.loc[group['MSE'].idxmin()]

best_per_model = grid_df.groupby('Model', group_keys=False).apply(get_best_row).reset_index(drop=True)
print("\n=== Best estimator per model ===")
print(best_per_model[['Model','n_estimators','C','gamma','MSE','MAE','R2']])


# 12) Overall best
best_row = best_per_model.loc[best_per_model['MSE'].idxmin()]
best_name = best_row['Model']
best_n = best_row.get('n_estimators', np.nan)
best_C = best_row.get('C', np.nan)
best_gamma = best_row.get('gamma', np.nan)
best_mdl = best_row['_mdl']

print(f"\n‚Üí Overall best model: {best_name} "
      f"{f'n_estimators={int(best_n)}' if pd.notna(best_n) else ''} "
      f"{f'C={best_C}' if pd.notna(best_C) else ''} "
      f"{f'gamma={best_gamma}' if pd.notna(best_gamma) else ''} "
      f"(MSE={best_row['MSE']:.4f}, MAE={best_row['MAE']:.4f}, R2={best_row['R2']:.4f})")

# 13) Forecast next 5 days using the best model
last = df.iloc[-1]

# Build future dates
future = pd.DataFrame({
    'Date': pd.date_range(start=last['Date'] + timedelta(days=1), periods=5, freq='D')
})

# Initialize numeric columns from the last row
future['Close']     = last['Close']
future['Volume']    = last['Volume']
future['sentiment'] = last['sentiment']

# Now apply the *same* feature engineering you did on df:

# ‚ë† Lag features
future['lag_Close']     = future['Close'].shift(1).fillna(last['Close'])
future['lag_Volume']    = future['Volume'].shift(1).fillna(last['Volume'])
future['lag_Sentiment'] = future['sentiment'].shift(1).fillna(last['sentiment'])

# ‚ë° Return & volume‚Äêchange
future['Daily_Return']  = future['Close'].pct_change().fillna(0)
future['Volume_Change'] = future['Volume'].pct_change().fillna(0)

# ‚ë¢ Moving averages
#   ‚Äì Use rolling with backfill so you don‚Äôt end up with NaNs at the front
future['MA_5']  = future['Close'].rolling(5).mean().bfill()
future['MA_10'] = future['Close'].rolling(10).mean().bfill()
future['MA_20'] = future['Close'].rolling(20).mean().bfill()

# ‚ë£ Volatility
future['Volatility_5']  = future['Daily_Return'].rolling(5).std().fillna(0)
future['Volatility_10'] = future['Daily_Return'].rolling(10).std().fillna(0)

# ‚ë§ Date‚Äêbased features
future['DayOfWeek'] = future['Date'].dt.dayofweek
future['Month']     = future['Date'].dt.month

# ‚ë• Drop any inf/nan just in case
future.replace([np.inf, -np.inf], np.nan, inplace=True)
future.fillna(0, inplace=True)

# ‚ë¶ Keep only the feature columns (dropping Date+Close)
future_X = future[features].astype(float)

# Predict
if best_name == 'SVR':
    future_X_scaled = scaler.transform(future_X)
    future['Predicted_Close'] = best_mdl.predict(future_X_scaled)
elif best_name == 'LSTM':
    # build sequences and take the last window
    seq_X, _ = create_lstm_sequences(pd.concat([X, future_X], ignore_index=True),
                                     pd.concat([y, pd.Series([0]*len(future_X))], ignore_index=True),
                                     window=lstm_window)
    future_seq = seq_X[-len(future_X):]
    future['Predicted_Close'] = best_mdl.predict(future_seq).flatten()
else:
    future['Predicted_Close'] = best_mdl.predict(future_X)

# 14) Plot actual vs forecast
one_month_ago = df['Date'].max() - pd.Timedelta(days=30)
recent = df[df['Date'] >= one_month_ago]

plt.figure(figsize=(10,6))
plt.plot(recent['Date'], recent['Close'],    marker='o', linestyle='-', label='Actual')
plt.plot(future['Date'],  future['Predicted_Close'],
         marker='s', linestyle='--', label='Forecast')
plt.title(f"{symbol} Close: Last 30d + Next 5d Forecast")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Step 0: Imports and Setup
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from newsapi import NewsApiClient
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Download VADER lexicon
nltk.download('vader_lexicon')

# Step 1: Fetch & prepare stock price data
symbol = "RELIANCE.NS"
df_price = yf.download(symbol, start="2023-01-01", end=datetime.today().strftime('%Y-%m-%d'))
df_price.columns = ['_'.join(c) if isinstance(c, tuple) else c for c in df_price.columns]
df_price = (
    df_price
    .reset_index()[['Date', f'Close_{symbol}', f'Volume_{symbol}']]
    .rename(columns={f'Close_{symbol}':'Close', f'Volume_{symbol}':'Volume'})
)
df_price['Date'] = pd.to_datetime(df_price['Date'])

# Step 2: Fetch & score news sentiment (last 30 days)
newsapi = NewsApiClient(api_key='7b20751a6c044d668d12c26c03f4f1d8')
to_date = datetime.today().strftime('%Y-%m-%d')
from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
arts = newsapi.get_everything(
    q="Reliance Industries",
    from_param=from_date,
    to=to_date,
    language='en',
    sort_by='relevancy',
    page_size=100
)['articles']
news_df = pd.DataFrame([{'date':a['publishedAt'][:10],'title':a['title']} for a in arts])
sid = SentimentIntensityAnalyzer()
news_df['sentiment'] = news_df['title'].map(lambda t: sid.polarity_scores(t)['compound'])
sentiment_daily = news_df.groupby('date')['sentiment'].mean().reset_index()
sentiment_daily['date'] = pd.to_datetime(sentiment_daily['date'])

# Step 3: Merge price + sentiment
df = (
    df_price
    .merge(sentiment_daily, how='left', left_on='Date', right_on='date')
    .drop(columns='date')
    .fillna(0)
    .sort_values('Date')
    .reset_index(drop=True)
)

# Step 4: Feature engineering
# 4.1 Lag Features
df['lag_Close'] = df['Close'].shift(1)
df['lag_Volume'] = df['Volume'].shift(1)
df['lag_Sentiment'] = df['sentiment'].shift(1)
# 4.2 Return & volume change
df['Daily_Return'] = df['Close'].pct_change()
df['Volume_Change'] = df['Volume'].pct_change()
# 4.3 Moving Averages
df['MA_5'] = df['Close'].rolling(5).mean()
df['MA_10'] = df['Close'].rolling(10).mean()
df['MA_20'] = df['Close'].rolling(20).mean()
# 4.4 Volatility
df['Volatility_5'] = df['Daily_Return'].rolling(5).std()
df['Volatility_10'] = df['Daily_Return'].rolling(10).std()
# 4.5 Date-based features
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['Month'] = df['Date'].dt.month

# Clean up
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

# Step 5: Prepare train/test
features = ['Volume','sentiment','lag_Close','lag_Volume','lag_Sentiment',
            'Daily_Return','Volume_Change','MA_5','MA_10','MA_20',
            'Volatility_5','Volatility_10','DayOfWeek','Month']
X, y = df[features], df['Close']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, shuffle=False)

# Step 6: Define models & hyperparameters
n_estimators_list = [50,100,200,300,500]
model_classes = {'RandomForest':RandomForestRegressor,
                 'XGBoost':XGBRegressor,'LightGBM':LGBMRegressor,
                 'GradientBoost':GradientBoostingRegressor}
def eval_model(mdl,X_t,y_t): return (mean_squared_error(y_t,mdl.predict(X_t)),
                                      mean_absolute_error(y_t,mdl.predict(X_t)),
                                      r2_score(y_t,mdl.predict(X_t)))
# SVR scaling
tn_scaler = StandardScaler()
X_train_scaled = tn_scaler.fit_transform(X_train)
X_test_scaled = tn_scaler.transform(X_test)
svr_params = {'C_list':[0.1,1,10],'gamma_list':[0.01,0.1]}

# Step 7: Prepare LSTM sequences
def create_lstm_sequences(X,y,window=1):
    Xs,ys=[],[]
    for i in range(len(X)-window): Xs.append(X.iloc[i:i+window].values); ys.append(y.iloc[i+window])
    return np.array(Xs),np.array(ys)
lstm_window=1
X_lstm,y_lstm=create_lstm_sequences(X,y,window=lstm_window)
split=int(len(X_lstm)*0.8)
X_lstm_train,X_lstm_test=X_lstm[:split],X_lstm[split:]
y_lstm_train,y_lstm_test=y_lstm[:split],y_lstm[split:]

# Step 8: Train & evaluate models
grid_results=[]
# tree-based
for name,cls in model_classes.items():
    for n in n_estimators_list:
        mdl=cls(n_estimators=n,random_state=1,verbosity=0) if name=='XGBoost' else cls(n_estimators=n,random_state=1)
        mdl.fit(X_train,y_train)
        mse,mae,r2=eval_model(mdl,X_test,y_test)
        grid_results.append([name,n,None,None,mse,mae,r2,mdl])
# SVR
for C in svr_params['C_list']:
    for g in svr_params['gamma_list']:
        svr=SVR(kernel='rbf',C=C,gamma=g); svr.fit(X_train_scaled,y_train)
        mse,mae,r2=eval_model(svr,X_test_scaled,y_test)
        grid_results.append(['SVR',None,C,g,mse,mae,r2,svr])
# LSTM
lstm_model=Sequential([LSTM(64,input_shape=(X_lstm_train.shape[1],X_lstm_train.shape[2])),Dense(1)])
lstm_model.compile(loss='mse',optimizer='adam')
lstm_model.fit(X_lstm_train,y_lstm_train,epochs=30,batch_size=8,verbose=0)
l_pred=lstm_model.predict(X_lstm_test).flatten()
grid_results.append(['LSTM',None,None,None,mean_squared_error(y_lstm_test,l_pred),
                     mean_absolute_error(y_lstm_test,l_pred),r2_score(y_lstm_test,l_pred),lstm_model])

# Step 9: Compile results & identify best
grid_df=pd.DataFrame(grid_results,columns=['Model','n_estimators','C','gamma','MSE','MAE','R2','_mdl'])
best_per=grid_df.groupby('Model',group_keys=False).apply(lambda g: g.loc[g['MSE'].idxmin()])
best=best_per.loc[best_per['MSE'].idxmin()]
best_mdl=best['_mdl']; best_name=best['Model']

# Step 10: Forecast next 5 days using best_mdl
last=df.iloc[-1]; future=pd.DataFrame({'Date':pd.date_range(start=last['Date']+timedelta(days=1),periods=5)})
# replicate features
for col in ['Close','Volume','sentiment']: future[col]=last[col]
# feature pipeline as before
def fea(df_tmp):
    df_tmp['lag_Close']=df_tmp['Close'].shift(1).fillna(last['Close'])
    df_tmp['lag_Volume']=df_tmp['Volume'].shift(1).fillna(last['Volume'])
    df_tmp['lag_Sentiment']=df_tmp['sentiment'].shift(1).fillna(last['sentiment'])
    df_tmp['Daily_Return']=df_tmp['Close'].pct_change().fillna(0)
    df_tmp['Volume_Change']=df_tmp['Volume'].pct_change().fillna(0)
    for m in [5,10,20]: df_tmp[f'MA_{m}']=df_tmp['Close'].rolling(m).mean().bfill()
    df_tmp['Volatility_5']=df_tmp['Daily_Return'].rolling(5).std().fillna(0)
    df_tmp['Volatility_10']=df_tmp['Daily_Return'].rolling(10).std().fillna(0)
    df_tmp['DayOfWeek']=df_tmp['Date'].dt.dayofweek; df_tmp['Month']=df_tmp['Date'].dt.month
fea(future)
# Handle NaNs after feature engineering
future.replace([np.inf, -np.inf], np.nan, inplace=True)
future.fillna(0, inplace=True)
X_fut=future[features]
if best_name=='SVR': future['Predicted_Close']=best_mdl.predict(tn_scaler.transform(X_fut))
elif best_name=='LSTM':
    seq,_=create_lstm_sequences(pd.concat([X,X_fut],ignore_index=True),pd.concat([y,pd.Series([0]*len(X_fut))]),window=lstm_window)
    future['Predicted_Close']=best_mdl.predict(seq[-len(X_fut):]).flatten()
else: future['Predicted_Close']=best_mdl.predict(X_fut)

# Step 11: Plot results
plt.figure(figsize=(10,6))
plt.plot(df['Date'].iloc[-30:],df['Close'].iloc[-30:],marker='o',label='Actual')
plt.plot(future['Date'],future['Predicted_Close'],marker='s',linestyle='--',label='Forecast')
plt.title(f"{symbol} Close: Last 30d + Next 5d Forecast")
plt.xlabel('Date');plt.ylabel('Close')
plt.xticks(rotation=45);plt.legend();plt.tight_layout();plt.show()

# Step 12: Additional Visuals
mask=df_price['Date']>=pd.Timestamp('2025-01-01')
fig,ax1=plt.subplots(figsize=(12,5))
ax1.plot(df_price.loc[mask,'Date'],df_price.loc[mask,'Close'],label='Close Price')
ax2=ax1.twinx(); ax2.bar(df_price.loc[mask,'Date'],df_price.loc[mask,'Volume'],alpha=0.3,label='Volume')
ax1.set_xlabel('Date');ax1.set_ylabel('Price');ax2.set_ylabel('Volume')
plt.title('Price & Volume (2025+)');plt.tight_layout();plt.show()

mask2=sentiment_daily['date']>=pd.Timestamp('2025-01-01')
plt.figure(figsize=(12,5))
plt.plot(sentiment_daily.loc[mask2,'date'],sentiment_daily.loc[mask2,'sentiment'],marker='o')
plt.title('News Sentiment (2025+)');plt.xlabel('Date');plt.ylabel('Sentiment');plt.tight_layout();plt.show()

mask3=df['Date']>=pd.Timestamp('2025-01-01')
corr=df.loc[mask3,features+['Close']].corr()
plt.figure(figsize=(10,8));sns.heatmap(corr,annot=True,center=0);plt.title('Feature Correlations (2025+)');plt.tight_layout();plt.show()

# End of script
```
